```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Dados_x = np.linspace(0,10,10)
Dados_y = Dados_x *Dados_x
plt.figure(figsize=(10,5), facecolor= 'gray')
Ax = plt.axes()
Ax.set_facecolor('white')
plt.plot(Dados_x,Dados_y,
color = 'red', linewidth=2,
markersize=5,
marker='o',
linestyle = 'dashed',
markeredgecolor = 'black')
plt.xlabel('Eixo x',fontweight='bold',
fontsize='large',fontfamily='fantasy',
color='blue')
plt.title('Grafico simples',fontweight='bold',
fontfamily='fantasy', color='blue')
plt.legend('numeros aleatorios',
loc = 'upper center', bbox_to_anchor=(1,0,0.0,
0.2), fontsize='large',shadow=True,borderpad=1)
plt.savefig("Grafico simples.jpg")
```

```python
plt.rcParams.update({'font.size':15})
plt.figure(figsize=(10,10))
linhas =2
colunas =2
plt.subplot(linhas,colunas,1)
plt.plot(Dados_x,'r--')
plt.title('Grafico 1')
#Grafico 2
plt.subplot(linhas,colunas,2)
plt.plot(Dados_y,'b--')
plt.title('Grafico 2')
#Grafico 3
plt.subplot(linhas,colunas,3)
plt.plot(Dados_y,'y--')
plt.title('Grafico 3')
#Grafico 4
plt.subplot(linhas,colunas,4)
plt.plot(Dados_y*3,'g--')
plt.title('Grafico 4')
plt.savefig("Graficos.jpg")
```

```python
df = pd.read_csv('monitoramento_tempo.csv',sep=',')
df.head(10)
```

```python
import datetime
df["data"]=pd.to_datetime(df['data'])
df.info()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['data'],df['temperatura'])
plt.title('temperatura x ano')
plt.xlabel('Ano')
plt.xlabel('Temperatura')
plt.legend('Temperatura')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'],
color='darkred',lw=0.7,linestyle='',marker='x')
eixo.set_title('temperatura no momento')
eixo.legend(['humidade'], loc='lower right',
fontsize=15)
eixo.set_xlabel('Ano',fontsize=20)
eixo.set_ylabel('Umidade',fontsize=20)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['humidade'],
color='pink', lw=0.7, linestyle='dashed')
eixo.set_title('umidade no momento')
eixo.legend(['humidade'], loc='lower right',
fontsize=15)
eixo.set_xlabel('Ano',fontsize=20)
eixo.grid(True)
eixo.set_ylabel('Umidade',fontsize=20)
```
